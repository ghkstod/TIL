## Groupby 연산
- SQL : Group By와 로직이 동일함
- groupby 기준컬럼: 패턴이 있는 문자열
- groupby 대상컬럼: 연속형 데이터
- 필요한 함수 : 집계함수(합, 평균, std 등)

```python
df.groupby('Location')['매출액'].agg(['sum','mean','std'])
```

```python
df.groupby('Location')['매출액'].agg(['sum','mean','std']).reset_index()
```

```python
speeds = pd.DataFrame(
    [
        ("bird", "Falconiformes", 389.0),
        ("bird", "Psittaciformes", 24.0),
        ("mammal", "Carnivora", 80.2),
        ("mammal", "Primates", np.nan),
        ("mammal", "Carnivora", 58),
    ],
    index=["falcon", "parrot", "lion", "monkey", "leopard"],
    columns=("class", "order", "max_speed"),
)
```

```python
speeds
```

```python
speeds.groupby("class").sum()
```

```python
speeds.groupby(["class", "order"]).sum()
```

```python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
df
```

```python
df.groupby("A")['C'].agg(['sum','mean','std']).reset_index()
```

```python
def get_letter_type(letter):
    if letter.lower() in 'aeiou':
        return 'vowel'
    else:
        return 'consonant'
```

```python
grouped = df.T.groupby(get_letter_type)
grouped
```

```python
lst = [1, 2, 3, 1, 2, 3]

s = pd.Series([1, 2, 3, 10, 20, 30], lst)

grouped = s.groupby(level=0)

grouped.first()
```

```python
df3 = pd.DataFrame({"X": ["A", "B", "A", "B"], "Y": [1, 4, 3, 2]})
df3.groupby(["X"]).get_group("A")
```

```python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
        "E":['F','F','T','T','F','T','F','F'],
        "F":[10,20,30,40,50,60,70,80]
    }
)
```

```python
df2 = df.set_index(["A", "B","E"])
```

```python
grouped = df2.groupby(level=df2.index.names.difference(["B","E"]))
```

```python
grouped.sum()
```

```python
grouped1 = df2.groupby('A')
```
