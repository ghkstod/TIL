## 데이터 통합하기
- 데이터 조인

```python
import numpy as np
import pandas as pd

df1=pd.DataFrame({
    'Class1':[95,92,98,100],
    'Class2':[91,93,97,99],
})

df2=pd.DataFrame({
    'Class1':[97,89],
    'Class2':[85,90]
})
```

```python
df1.append(df2) #공식문서 확인
# https://pandas.pydata.org/pandas-docs/version/1.4/reference/api/pandas.DataFrame.append.html
```

```python
res=pd.concat([df1,df2])
res
```

```python
df3=pd.DataFrame({
    'Class1':[96,83]
})
df3
```

```python
res1=pd.concat([res,df3],ignore_index=True)
res1
```

## merge()

```python
df_A_B = pd.DataFrame({'판매월': ['1월', '2월', '3월', '4월'],
                       '제품A': [100, 150, 200, 130],
                       '제품B': [90, 110, 140, 170]})

df_C_D = pd.DataFrame({'판매월': ['1월', '2월', '3월', '4월'],
                       '제품C': [112, 141, 203, 134],
                       '제품D': [90, 110, 140, 170]})

df_C_D
```

```python
df_A_B.merge(df_C_D)
```

```python
df_A_B.join(df_C_D) #동일한 컬럼이 존재해 에러가 발생
```

```python
df_left = pd.DataFrame({'key':['A','B','C'], 'left': [1, 2, 3]})
df_right = pd.DataFrame({'key':['A','B','D'], 'right': [4, 5, 6]})
```

```python
df_left.merge(df_right,how='outer',on='key') #전체
```

```python
df_left.merge(df_right,how='right',on='key') #key가 오른쪽 기준
```

```python
df_left.merge(df_right,how='left',on='key') #key가 왼쪽 기준
```

```python
df_left.merge(df_right,how='inner',on='key') # key공통
```
